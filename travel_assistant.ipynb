```python

def voice_chat(user_voice):

    messages = [
    {"role": "system", "content": "You are a kindful ai travel assistant. Help user with their travel and tourist works and asssist them by solving their doubts "},
    ]
    user_message = model.transcribe(user_voice,language='en')["text"]
    print(user_message)
    messages.append(
        {"role": "user", "content": user_message},
    )



    chat = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", messages=messages,max_tokens=300
    )

    reply = chat.choices[0].message.content
    print(reply)

    messages.append({"role": "assistant", "content": reply})

    tts.tts_to_file(text=reply, file_path="output.wav")

    return(reply, 'output.wav')

```

```python
text_reply = gr.Textbox(label="Generated Text")
voice_reply = gr.Audio('output.wav')

gr.Interface(
    title = 'AI Voice Assistant by SHA',
    fn=voice_chat,
    inputs=[gr.Audio(sources=["microphone"],type="filepath")],
    outputs=[
        text_reply,  voice_reply
    ]).launch(debug=True)
```
